## SGHEAD

```python
mode = "a"
interest = "og"
htype = "sghead"
r=1
dsdcts_dir = f"{cwd}/inputs/{mode}/{htype}_dsdcts"
models_dir = f"{cwd}/models/{mode}/{htype}/{interest}"
results_dir = f"{cwd}/results/{mode}/{htype}/{interest}"
quant = True
model_name = "microsoft/deberta-v3-base"
results_file = f"{results_dir}/randp_{model_name.split("/")[-1]}_{r}.json"
#s_metrics, s_preds, s_reals, s_inputids = evaluate_model(mode, htype, model_name, models_dir, dsdcts_dir, r, quant=quant)
```

## mhead

```python
mode = "a"
interest = "og"
htype = "mhead"
r=1
dsdcts_dir = f"{cwd}/inputs/{mode}/{htype}_dsdcts"
models_dir = f"{cwd}/models/{mode}/{htype}/{interest}"
results_dir = f"{cwd}/results/{mode}/{htype}/{interest}"
quant = True
model_name = "microsoft/deberta-v3-base"
results_file = f"{results_dir}/randp_{model_name.split("/")[-1]}_{r}.json"
#m_metrics, m_preds, m_reals, m_inputids = evaluate_model(mode, htype, model_name, models_dir, dsdcts_dir, r, quant=quant)
```

```python
with open(results_file, "r", encoding="utf-8") as f:
    randp = json.load(f)
heads = list(randp)
m_reals = {head:[] for head in heads}
m_preds = {head:[] for head in heads}
m_inputs = []
for head in heads:
    for art in randp[head]:
        inputs = [ent[0] for ent in art]
        reals = [ent[1] for ent in art]
        preds = [ent[2] for ent in art]
        m_reals[head].append(reals)
        m_preds[head].append(preds)
        if head == heads[0]:
            m_inputs.append(inputs)
```

```python
for sentid in range(len(m_reals["Actor"][:5])):
    # check effects of entity length on accuracy
    print("\n", sentid)
    print(f"{'':15s}Act\tInst\tObj\tRes\tTime")
    for tokid in range(len(m_reals["Actor"][sentid])):
        prnt = f"{m_inputs[sentid][tokid]:15s}"
        for lbl in list(m_reals):
            prnt += f"{m_reals[lbl][sentid][tokid]} {m_preds[lbl][sentid][tokid]}\t"
        print(prnt)
```

#### Dice

DSC = 2|x{intersection}y|/(|x|+|y|)

DSC = 2J/(1+J)

#### Jaccard

J(A,B) = |A{intersection}B|/(|A|+|B|-|A{intersection}B|)

JC = TP/(TP+FP+FN)

J= DSC/(2-DSC)

```python
s_reals_flat = [tok for sent in s_reals for tok in sent]
s_preds_flat = [tok for sent in s_preds for tok in sent]

for avg_method in ["micro", "micro_ent", "weighted", "weighted_ent", "macro", "macro_ent"]:
    if "ent" in avg_method:
        avg_method = avg_method.split("_")[0]
        entity_labels = [label for label in list(set(s_reals_flat)) if label != 'O']
    else:
        entity_labels = list(set(s_reals_flat))
    s_dice = f1_score(s_reals_flat, s_preds_flat, average=avg_method, labels=entity_labels)
    s_jaccard = jaccard_score(s_reals_flat, s_preds_flat, average=avg_method, labels=entity_labels)
    print(f"{avg_method:10s}Dice: {round(s_dice*100,2)}\tJaccard: {round(s_jaccard*100,2)}")
    
```
